```python
gsheet_name = "test_upload"
tab_name = 'weight'

def get_gsheet_data(gsheet_name, tab_name):
    gc = gspread.service_account(filename='gsheet_creds/football-373419-fe3d09a53ac9.json')
    sh = gc.open(gsheet_name) 
    worksheet = sh.worksheet(tab_name)
    df = pd.DataFrame(worksheet.get_all_records())
    
    return df
```

```python
data = get_gsheet_data(gsheet_name, tab_name)
data.head()
```

```python
heights = pd.DataFrame(np.random.randn(10, 4), columns=list('ABCD'))
heights.head()
```

```python
gsheet_name = "test_upload"
tab_name = "height"

def write_df_to_gsheets(gsheet_name, tab_name, df):
    gc = gspread.service_account(filename='gsheet_creds/football-373419-fe3d09a53ac9.json')
    sh = gc.open(gsheet_name) 
    worksheet = sh.worksheet(tab_name)
    set_with_dataframe(worksheet, df)
```

```python
write_df_to_gsheets(gsheet_name, tab_name, heights)
```

```python
match_importance = pd.read_csv("../data/match_importance.csv", index_col=0)
sim_season = pd.read_csv("../data/simulated_season.csv", index_col=0)
```

```python
gsheet_name = "test_upload"
tab_name = "match_importance"
df_upload = match_importance

write_df_to_gsheets(gsheet_name, tab_name, df_upload)
```
